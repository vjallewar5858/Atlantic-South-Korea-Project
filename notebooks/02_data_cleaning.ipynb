```python
import pandas as pd
import numpy as np

df = pd.read_csv('Atlantic_South_Korea.csv')
print("Raw data loaded:", df.shape)
```

```python
#converting date into datetime
df['date'] = pd.to_datetime(df['date'], dayfirst=True)

# Verify
print(df['date'].dtype)
print(df['date'].head())
```

```python
#converting duration into minutes
df['duration_min'] = (df['duration_ms'] / 60000).round(2)
df.drop(columns=['duration_ms'], inplace=True)

# Verify
print(df['duration_min'].describe())
```

```python
#droping album_cover_url as it is not necessary
df.drop(columns=['album_cover_url'], inplace=True)

print("Columns remaining:", df.columns.tolist())
```

```python
#remove duplicates 
print("Before:", df.shape)
df.drop_duplicates(inplace=True)
print("After:", df.shape)
print("Removed:", 27800 - len(df), "duplicate rows")
```

```python
#reset Index
df.reset_index(drop=True, inplace=True)
print("Index reset done")
```

```python
print("Shape:", df.shape)
print("\nDtypes:\n", df.dtypes)
print("\nNulls:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
df.head()
```

```python
df.to_csv('cleaned_data.csv', index=False)
print("Cleaned data saved!")
```
